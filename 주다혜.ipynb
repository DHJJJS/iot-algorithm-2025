- 1. P511 배낭함수 문제의 1, 2에 적합한 코드를 골라서 작성하시오.  

```python
def knapsack():
    array = [[0 for _ in range(maxWeight+1)] for _ in range(rowCount+1)] # 테이블을 만드는 코드
    for row in range(1, rowCount+1): # 배낭을 1개부터 넣는다
        for col in range(1, maxWeight+1): # 배낭의 무게가 1부터 채워진다
                if weight[row] > col:    ## 이전 값보다 컬럼 (무게)가 작으면
                    array[row][col] = array[row-1][col]  ## 이전값을 그대로 가지고 내려오고 # 현재 값의 최대 가치가 허용 무게보다 크다면, 이전 값의 최대 가치를 그대로 사용
                else:
                    value1 = money[row] + array[row-1][col-weight[row]] ## 최대무게에서 현재들어갈 무게를 뺀 위치의 가치+ 현재보석의 가격 # 값 1은 현재 값의 가치 + 이전 값의 허용 무게 - 이전 값의 무게 = 이전 값의 남는 최대 가치
                    value2 = array[row-1][col]  ## 이전위치 값 # 값 2은 이전값의 최대 허용 가치
                    array[row][col] = max(value1, value2) ## 둘중에 비싼 걸 # 현재값의 최대 허용 가치는 값1과 값2의 최대값

    return array[rowCount][maxWeight] ## 최종 행렬 마지막값(array[4][7])을 리턴

maxWeight = 7
rowCount = 4
weight = [0, 6, 4, 3, 5]
money = [0, 13, 8, 6, 12]

maxValue = knapsack()
print(f'배낭문제 최대 보석 가격 = {maxValue}')
```

-  2. P269 유명 맛집 대기줄 구현하기 문제를 while True:문을 사용하여 동적으로 입력을 받는 프로그램으로 변경작성하시오.

```python
def isQueueFull():
    global SIZE, queue, front, rear # 전역변수를 함수 내에서 변경하고자 할 때 선언하는 global

    if rear != SIZE - 1:
        return False
    elif rear == SIZE - 1 and front == -1:
        return True
    else:
        for i in range(front+1, SIZE):
            queue[i-1] = queue[i] # 데이터 앞 빈자리에 첫번째 데이터 옮김
            queue[i] = None

        front -= 1
        rear -= 1
        return False
    
def isQueueEmpty():
    global SIZE, queue, front, rear
    if front == rear:
        return True
    else:
        return False

def enQueue(data):
    global SIZE, queue, front, rear
    if isQueueFull():
        print('Queue is full')
    else:
        rear += 1
        queue[rear] = data

def deQueue():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('Queue is Empty')
        return None
    else:
        front += 1
        data = queue[front]
        queue[front] = None
        return data

def peek():
    global SIZE, queue, front, rear
    if isQueueEmpty():
        print('Queue is empty')
        return None
    else:
        return queue[front+1]
    
# 초기화
SIZE = 5
queue = [None for _ in range(SIZE)]
front = rear = -1

## 메인 모듈
if __name__ == '__main__':
    while True:
        select = input('삽입(I)/추출(E)/확인(V)/종료(Q) --> ').upper()

        if select == 'Q':
            break
        elif select == 'I':
            data = input('데이터 입력 > ')
            enQueue(data)
            print(f'대기줄 상태 = {queue}')
        elif select == 'E':
            data = deQueue()
            print(f'손님입장 : {data}')
            print(f'대기줄 상태 : {queue}')
        elif select == 'V':
            data = peek()
            print(f'대기자명 : {data}')
            print(f'대기자 상태 : {queue}')
        else:
            print('입력 오류')
```

- 3. P309 편의점 판매된 물건 목록 출력하기 문제 역시 while True: 문으로 동적으로 입력받는 프로그램으로 변경하시오.
데이터의 종류는 [ 콘말차, 삿포로 맥주, 아카페라 벤티 헤이즐넛, 레어치즈푸딩, 척오리지널 메가 사워, 요아정 요거트컵, 페퍼로니피자주먹밥, 널담 슈톨렌, 딸기마시멜로케이크, 버니공쥬 딸기뚱카롱, 고추잡채호빵, 체다 슈크림붕어스낵 ]

```python
def isStackFull():   # 스택이 꽉 찻는지 확인하는 함수
    global SIZE, stack, top
    if (top == SIZE - 1): # Full
        return True
    else:
        return False
    
def isStackEmpty():
    global SIZE, stack, top
    if (top == -1):
        return True
    else:
        return False
    
def push(data): # 스택에 데이터 추가
    global SIZE, stack, top
    if isStackFull(): # isStackFull() == True와 동일
        print('Stack is full')
        return
    else:
        top += 1
        stack[top] = data

def pop(): # 스택에서 데이터 추출
    global SIZE, stack, top
    if isStackEmpty():
        print('Stack is empty.')
    else:
        data = stack[top]
        stack[top] = None
        top -= 1
        return data

def peek(): # 스택의 top위치의 데이터 확인(살짝보기)
    global SIZE, stack, top
    if isStackEmpty():
        print('Stack is empty.')
        return None
    else:
        return stack[top]

# 반복문에서 사용되지 않는 변수(i)는 _로 변경 가능
cafes = ['콘말차', '삿포로 맥주', '아카페라 벤티 헤이즐넛', '레어치즈푸딩', '척오리지널 메가 사워', '요아정 요거트컵', '페퍼로니피자주먹밥', '널담 슈톨렌', '딸기마시멜로케이크', '버니공쥬 딸기뚱카롱', '고추잡채호빵', '체다 슈크림붕어스낵']
SIZE = len(cafes)
stack = [None for _ in range(SIZE)]
top = -1

if __name__ == '__main__':

    while True:
        select = input('삽입(I)/추출(E)/확인(V)/종료(Q) > ').upper()
        
        if select == 'Q':
            break
        elif select == 'I':
            data = input('입력 데이터 > ')
            push(data)
            print(f'스택상태\n{stack}')
        elif select == 'E':
            data = pop()
            print(f'추출 데이터 > {data}')
            print(f'스택상태\n{stack}')
        elif select == 'V':
            data = peek()
            print(f'확인 데이터 > {data}')
            print(f'스택상태\n{stack}')
        else:
            print('입력오류~!')

## 이진트리로 푼다
```

- 4. P473 Self Study 13-2 의 문제. 이진 검색 알고리즘을 풀면서 검색 횟수도 같이 출력하도록 구현하세요.

```python
import random

def binSearch(ary, fData):
    pos = -1
    start = 0
    count = 0
    end = len(ary) - 1

    while (start <= end):
        mid = (start + end) // 2
        if fData == ary[mid]:
            count += 1
            return mid
        elif fData > ary[mid]:
            start = mid + 1
            count += 1
        else:
            end = mid - 1
            count += 1

        print(f'검색 종료 : 총 검색횟수 {count}')

    return pos

dataAry = [random.randint(0, 1000000) for _ in range(10000)]
findData = int(input('찾을 수를 입력하세요 > '))

print('배열 ->', dataAry)
position = binSearch(dataAry, findData)
if position == -1:
    print(findData, '(이)가 없네요.')
else:
    print(findData, '(은)는 ', position, '위치에 있음')
```
