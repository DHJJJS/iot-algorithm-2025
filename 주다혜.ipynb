- 1. P511 배낭함수 문제의 1, 2에 적합한 코드를 골라서 작성하시오.  

```python
def knapsack():
    array = [[0 for _ in range(maxWeight+1)] for _ in range(rowCount+1)] # 테이블을 만드는 코드
    for row in range(1, rowCount+1): # 배낭을 1개부터 넣는다
        for col in range(1, maxWeight+1): # 배낭의 무게가 1부터 채워진다
                if weight[row] > col:
                    array[row][col] = array[row-1][col] # 현재 값의 최대 가치가 허용 무게보다 크다면, 이전 값의 최대 가치를 그대로 사용
                else:
                    value1 = money[row] + array[row-1][col-weight[row]] # 값 1은 현재 값의 가치 + 이전 값의 허용 무게 - 이전 값의 무게 = 이전 값의 남는 최대 가치
                    value2 = array[row-1][col] # 값 2은 이전값의 최대 허용 가치
                    array[row][col] = max(value1, value2) # 현재값의 최대 허용 가치는 값1과 값2의 최대값

    return array[rowCount][maxWeight]
```

-  2. P269 유명 맛집 대기줄 구현하기 문제를 while True:문을 사용하여 동적으로 입력을 받는 프로그램으로 변경작성하시오.

```python
def isStackFull():   # 스택이 꽉 찻는지 확인하는 함수
    global SIZE, stack, top
    if (top == SIZE - 1): # Full
        return True
    else:
        return False
    
def isStackEmpty():
    global SIZE, stack, top
    if (top == -1):
        return True
    else:
        return False
    
def push(data): # 스택에 데이터 추가
    global SIZE, stack, top
    if isStackFull(): # isStackFull() == True와 동일
        print('Stack is full')
        return
    else:
        top += 1
        stack[top] = data

def pop(): # 스택에서 데이터 추출
    global SIZE, stack, top
    if isStackEmpty():
        print('Stack is empty.')
    else:
        data = stack[top]
        stack[top] = None
        top -= 1
        return data

def peek(): # 스택의 top위치의 데이터 확인(살짝보기)
    global SIZE, stack, top
    if isStackEmpty():
        print('Stack is empty.')
        return None
    else:
        return stack[top]

# 반복문에서 사용되지 않는 변수(i)는 _로 변경 가능
SIZE = int(input('스택크기를 결정하세요 --> '))
stack = [None for _ in range(SIZE)]
top = -1

if __name__ == '__main__':

    while True:
        select = input('삽입(I)/추출(E)/확인(V)/종료(Q) > ').upper()
        
        if select == 'Q':
            break
        elif select == 'I':
            data = input('입력 데이터 > ')
            push(data)
            print(f'스택상태\n{stack}')
        elif select == 'E':
            data = pop()
            print(f'추출 데이터 > {data}')
            print(f'스택상태\n{stack}')
        elif select == 'V':
            data = peek()
            print(f'확인 데이터 > {data}')
            print(f'스택상태\n{stack}')
        else:
            print('입력오류~!')
```

- 4. P473 Self Study 13-2 의 문제. 이진 검색 알고리즘을 풀면서 검색 횟수도 같이 출력하도록 구현하세요.

```python
def binSearch(ary, fData):
    pos = -1
    start = 0
    count = 0
    end = len(ary) - 1

    while (start <= end):
        mid = (start + end) // 2
        if fData == ary[mid]:
            count += 1
            return mid
        elif fData > ary[mid]:
            start = mid + 1
            count += 1
        else:
            end = mid - 1
            count += 1

        print(f'검색 종료 : 총 검색횟수 {count}')

    return pos

dataAry = [50, 60, 105, 120, 150, 160, 162, 168, 177, 188]
findData = 162 # 할머니 키

print('배열 ->', dataAry)
position = binSearch(dataAry, findData)
if position == -1:
    print(findData, '(이)가 없네요.')
else:
    print(findData, '(은)는 ', position, '위치에 있음')
```
